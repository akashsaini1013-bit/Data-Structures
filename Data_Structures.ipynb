4. Describe how dictionaries store data?
- Python dictionaries store data as unordered collections of key-value pairs using a
hash table implementation for fast lookups. Keys must be immutable and hashable,
enabling average O(1) access time, while values can be any data type.
5.Why might you use a set instead of a list in Python?
- Sets provide unique elements and constant-time lookups, outperforming lists when
duplicates must be avoided or membership checks are frequent. Lists preserve order
and allow indexing but scan linearly for checks, slowing down with size.
6. What is a string in Python, and how is it different from a list?
- Strings in Python are immutable sequences of Unicode characters, created with
quotes like s = "hello". They differ from lists, which are mutable sequences of any data
types, defined with brackets like my_list = [1, "a", True] .
7.How do tuples ensure data integrity in Python?
- Tuples ensure data integrity in Python primarily through their immutability, preventing
accidental or unintended modifications once created. This fixed nature protects data
from changes across functions, threads, or program execution.
8. What is a hash table, and how does it relate to dictionaries in Python?
- Hash tables provide fast key-value storage using a hash function to map keys to array
indices, enabling average O(1) lookups, inserts, and deletes. Python dictionaries
implement hash tables internally, leveraging this for efficient operations on unordered
collections.
Python Dictionary Relation
Dictionaries (dict) are hash tables with keys as immutable hashable objects (strings,
tuples) and values as any type. Internally, CPython uses a resizable array of entries
(hash, key, value tuples) plus an indices array, resizing when load factor hits ~2/3 for
performance . Since Python 3.6, they maintain insertion order.

even other lists.
